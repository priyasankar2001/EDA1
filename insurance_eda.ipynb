### Q6: What is the distribution of premium?

```python
sns.set_theme(style="ticks", palette="pastel")
sns.histplot(x='premium', data=df,kde=True)
plt.title("distribution of premium")
```

### Q7: Is the distribution of premium skewed? If yes, in which direction?

**Answer:** Yes, the premium distribution is **right-skewed (positively skewed)**. Most insurance premiums are concentrated at lower values, with a long tail extending toward higher premiums. This is typical for insurance data, where a small percentage of high-risk individuals (especially smokers) pay significantly higher premiums.
